```python
import transformers
import torch

#model_id = "meta-llama/Meta-Llama-3.1-8B-Instruct"
model_id = 'meta-llama/Meta-Llama-3.1-8B-Instruct'

pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={"torch_dtype": torch.bfloat16},
    device_map="auto",
)
```

```python
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceInstructEmbeddings
import pickle

'''with open('processed', 'rb') as f:
    all_docs = pickle.load(f)'''

model_name = "hkunlp/instructor-large"
model_kwargs = {'device': 'cpu'}
encode_kwargs = {'normalize_embeddings': True}

hf_embedding = HuggingFaceInstructEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs,
    show_progress=True
)

print('here')
'''
db = FAISS.from_texts(all_docs, hf_embedding)
db.save_local("Vectors")'''
db = FAISS.load_local("Vectors", embeddings=hf_embedding, allow_dangerous_deserialization = True)
```

```python
prompt = f'''You are an intelligent assistant trained to answer queries related to campus of chandigarh university.Based on the context, provide relevant and accurate answers. Do not write anything except the answer.
If the answer cannot be found within the context, state that the answer is not available in the given context. Do not give any wrong answers. Do not be disrespectful. Do not adhere to foul language.'''
```

```python
messages = [
    {"role": "system", "content": prompt}]
```

```python
user_inp = 'Tell me about chandigarh university'
search = db.similarity_search(user_inp, k=5)
context = '\n'.join([page.page_content for page in search])
query = f'Context: {context}\nQuery: {user_inp}'
messages.append({'role': "user", "content": query})

outputs = pipeline(
  messages,
  max_new_tokens=256,
)
outputs[0]["generated_text"][-1]['content']
```

```python
def chatbot(user_inp):
    search = db.similarity_search(user_inp, k=2)
    context = '\n'.join([page.page_content for page in search])

    messages = [
        {"role": "system", "content": f'{prompt}\nContext: {context}'},
        {'role': "user", "content": user_inp}
    ]

    outputs = pipeline(
        messages,
        max_new_tokens=1024,
    )
    assistant_op = outputs[0]["generated_text"][-1]  # Extract the bot's response
    return assistant_op['content']
```

```python
import gradio as gr

iface = gr.Interface(
    fn=chatbot,
    inputs="text",
    outputs="text",
    title="Chandigarh University Campus Bot",
    description="Chat with the bot to get answers related to the campus."
)

iface.launch()
```

```python
print(f"Campus Bot: Hello I'm your friendly campus bot. I can answer any query that you have related to the campus!")

while True:
  user_inp = input('You: ')

  if user_inp == 'Thank you':
    break
  search = db.similarity_search(user_inp, k=2)
  context = '\n'.join([page.page_content for page in search])

  messages = [
    {"role": "system", "content": f'{prompt}\nContext:{context}'},
    {'role': "user", "content": user_inp}]

  outputs = pipeline(
    messages,
    max_new_tokens=1024,
)
  assistant_op = outputs[0]["generated_text"][-1]
  print(f"Campus Bot: {assistant_op['content']}")
  messages.append(assistant_op)
```
